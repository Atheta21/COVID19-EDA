```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
top_confirmed = data.sort_values("Confirmed", ascending=False).head(10)
plt.figure()
sns.barplot(data=top_confirmed, x="Confirmed", y="Country/Region", palette="Reds_r")
plt.title("Top 10 Countries by Confirmed Cases")
plt.show()
```

```python
plt.figure()
sns.histplot(data["Deaths / 100 Cases"], bins=20, kde=True, color="red")
plt.title("Distribution of Deaths per 100 Cases")
plt.show()
```

plot fig, plot histogram and use death/100 case data, bins=20 means Divides the range of data into 20 intervals (bars).
kde=True--Adds a kernel density estimate (a smooth curve over the histogram).
color="red"--Makes both the histogram and KDE curve red.

```python
plt.figure()
sns.histplot(data["Recovered / 100 Cases"], bins=20, kde=True, color="green")
plt.title("Distribution of Recovered per 100 Cases")
plt.show()
```

```python
region_grouped = data.groupby("WHO Region").sum(numeric_only=True).sort_values("Confirmed", ascending=False)
plt.figure()
sns.barplot(x=region_grouped["Confirmed"], y=region_grouped.index, palette="Blues_d")
plt.title("WHO Region-wise Confirmed Cases")
plt.show()
```

```python
corr = data.select_dtypes(include=["float64", "int64"]).corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

corr()=preparing the data to visualize which numeric columns are strongly related (e.g., Confirmed â†” Deaths) .corr--The correlation matrix to plot (2D table of values).annot=True--Displays the actual correlation values inside each square.cmap="coolwarm"	A color map: blue = negative correlation, red = positive. linewidths=0.5--Adds thin lines between squares to improve visibility.
